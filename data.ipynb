```python
# Diabetes Data Preprocessing - Complete Pipeline
# All phases in one continuous Jupyter Notebook
# Fixed version without imblearn dependency

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.decomposition import PCA
from sklearn.impute import KNNImputer
from scipy import stats
import missingno as msno
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('default')
sns.set_palette("husl")

print("‚úÖ All libraries imported successfully!")

# =============================================================================
# PHASE 1: DATA COLLECTION & UNDERSTANDING
# =============================================================================

print("=" * 60)
print("PHASE 1: DATA COLLECTION & UNDERSTANDING")
print("=" * 60)

# Load the dataset
try:
    df = pd.read_csv('diabetes.csv')
    print("üìä Dataset loaded successfully from 'diabetes.csv'")
except FileNotFoundError:
    print("‚ö†Ô∏è File not found. Creating sample dataset for demonstration...")
    np.random.seed(42)
    n_samples = 1000
    data = {
        'Pregnancies': np.random.randint(0, 15, n_samples),
        'Glucose': np.random.normal(120, 30, n_samples),
        'BloodPressure': np.random.normal(70, 12, n_samples),
        'SkinThickness': np.random.normal(20, 10, n_samples),
        'Insulin': np.random.normal(80, 50, n_samples),
        'BMI': np.random.normal(25, 5, n_samples),
        'DiabetesPedigreeFunction': np.random.exponential(0.5, n_samples),
        'Age': np.random.randint(21, 80, n_samples),
        'Outcome': np.random.randint(0, 2, n_samples)
    }
    df = pd.DataFrame(data)
    print("‚úÖ Sample dataset created successfully!")

# Basic information
print(f"\nüìê Dataset Shape: {df.shape}")
print(f"üìù Columns: {list(df.columns)}")
print(f"\nüîß Data Types:\n{df.dtypes}")

# Display first few rows
print("\nüëÄ First 5 rows:")
display(df.head())

# Basic statistics
print("\nüìà Basic Statistics:")
display(df.describe())

# Identify Data Quality Issues
print("\n" + "=" * 60)
print("DATA QUALITY ISSUES IDENTIFICATION")
print("=" * 60)

# 1. Biological Impossibilities (Zeros)
zero_sensitive_features = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
print("üîç Biological Impossibilities (Zeros):")
zero_summary = {}
for feature in zero_sensitive_features:
    zero_count = (df[feature] == 0).sum()
    percentage = (zero_count / len(df)) * 100
    zero_summary[feature] = zero_count
    if zero_count > 0:
        print(f"  ‚ùå {feature}: {zero_count} zeros ({percentage:.2f}%)")
    else:
        print(f"  ‚úÖ {feature}: No zeros found")

# 2. Data Type Inconsistencies
print(f"\nüìä Data Type Consistency: All features are {df.dtypes.unique()}")

# 3. Potential Outliers
print("\nüìä Potential Outliers (IQR Method):")
numerical_features = df.select_dtypes(include=[np.number]).columns.drop('Outcome', errors='ignore')
outlier_summary = {}
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(3, 3, i)
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    outlier_summary[feature] = len(outliers)
    sns.boxplot(y=df[feature])
    plt.title(f'{feature}\n({len(outliers)} outliers)')
    print(f"  üìå {feature}: {len(outliers)} outliers")
plt.tight_layout()
plt.show()

# =============================================================================
# PHASE 2: DATA CLEANING
# =============================================================================

print("\n" + "=" * 60)
print("PHASE 2: DATA CLEANING")
print("=" * 60)

# Create a copy for cleaning
df_clean = df.copy()
print("‚úÖ Created clean copy of the dataset")

# 1. Handle Impossible Zeros
print("\n1. üö´ HANDLING IMPOSSIBLE ZEROS:")
zero_sensitive_features = {
    'Glucose': (0, "Glucose cannot be 0"),
    'BloodPressure': (0, "Blood pressure cannot be 0"),
    'SkinThickness': (0, "Skin thickness cannot be 0"),
    'Insulin': (0, "Insulin cannot be 0"),
    'BMI': (0, "BMI cannot be 0")
}
zeros_replaced = 0
for feature, (threshold, reason) in zero_sensitive_features.items():
    zero_count = (df_clean[feature] == threshold).sum()
    if zero_count > 0:
        print(f"  üîÑ {feature}: Replacing {zero_count} zeros with NaN")
        df_clean[feature] = df_clean[feature].replace(threshold, np.nan)
        zeros_replaced += zero_count
print(f"‚úÖ Total zeros replaced with NaN: {zeros_replaced}")

# 2. Missing Values Analysis
print("\n2. üìä MISSING VALUES ANALYSIS:")
plt.figure(figsize=(12, 6))
msno.matrix(df_clean)
plt.title('Missing Values Matrix After Zero Replacement')
plt.show()
missing_percent = (df_clean.isnull().sum() / len(df_clean)) * 100
print("\nMissing Value Percentages:")
missing_data = pd.DataFrame({
    'Feature': missing_percent.index,
    'Missing_Percentage': missing_percent.values,
    'Missing_Count': df_clean.isnull().sum().values
})
missing_data = missing_data[missing_data['Missing_Count'] > 0]
if len(missing_data) > 0:
    display(missing_data)
else:
    print("  ‚úÖ No missing values found!")

# 3. Imputation Strategy
print("\n3. üîß IMPUTATION STRATEGY:")
imputer = KNNImputer(n_neighbors=5)
features_to_impute = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df_clean[features_to_impute] = imputer.fit_transform(df_clean[features_to_impute])
print("  ‚úÖ Applied KNN Imputer (n_neighbors=5)")

# 4. Outlier Treatment
print("\n4. üìè OUTLIER TREATMENT:")
for feature in numerical_features:
    Q1 = df_clean[feature].quantile(0.25)
    Q3 = df_clean[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers_before = ((df_clean[feature] < lower_bound) | (df_clean[feature] > upper_bound)).sum()
    if outliers_before > 0:
        df_clean[feature] = np.clip(df_clean[feature], lower_bound, upper_bound)
        outliers_after = ((df_clean[feature] < lower_bound) | (df_clean[feature] > upper_bound)).sum()
        print(f"  üîß {feature}: Capped {outliers_before} outliers ‚Üí {outliers_after} remaining")
print("\n‚úÖ Data cleaning completed!")
print(f"üìä Final dataset shape: {df_clean.shape}")

# =============================================================================
# PHASE 3: DATA TRANSFORMATION
# =============================================================================

print("\n" + "=" * 60)
print("PHASE 3: DATA TRANSFORMATION")
print("=" * 60)

# 1. Feature Engineering
print("1. üõ†Ô∏è FEATURE ENGINEERING:")
bins_age = [20, 35, 50, 65, 100]
labels_age = ['Young', 'Middle-aged', 'Senior', 'Elderly']
df_clean['AgeGroup'] = pd.cut(df_clean['Age'], bins=bins_age, labels=labels_age, right=False)
bins_bmi = [0, 18.5, 25, 30, 100]
labels_bmi = ['Underweight', 'Normal', 'Overweight', 'Obese']
df_clean['BMICategory'] = pd.cut(df_clean['BMI'], bins=bins_bmi, labels=labels_bmi, right=False)
bins_glucose = [0, 70, 100, 125, 200, 300]
labels_glucose = ['Low', 'Normal', 'Prediabetes', 'Diabetes', 'Severe Diabetes']
df_clean['GlucoseCategory'] = pd.cut(df_clean['Glucose'], bins=bins_glucose, labels=labels_glucose, right=False)
print("  ‚úÖ Created features: AgeGroup, BMICategory, GlucoseCategory")

# Display new features summary
print("\nüìã New Features Summary:")
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
age_counts = df_clean['AgeGroup'].value_counts()
axes[0].bar(age_counts.index, age_counts.values, color='skyblue')
axes[0].set_title('Age Group Distribution')
axes[0].tick_params(axis='x', rotation=45)
bmi_counts = df_clean['BMICategory'].value_counts()
axes[1].bar(bmi_counts.index, bmi_counts.values, color='lightgreen')
axes[1].set_title('BMI Category Distribution')
axes[1].tick_params(axis='x', rotation=45)
glucose_counts = df_clean['GlucoseCategory'].value_counts()
axes[2].bar(glucose_counts.index, glucose_counts.values, color='lightcoral')
axes[2].set_title('Glucose Category Distribution')
axes[2].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
print(f"Age Groups: {dict(age_counts)}")
print(f"BMI Categories: {dict(bmi_counts)}")
print(f"Glucose Categories: {dict(glucose_counts)}")

# 2. Encoding
print("\n2. üî† ENCODING:")
label_encoder = LabelEncoder()
ordinal_features = ['AgeGroup', 'BMICategory', 'GlucoseCategory']
for feature in ordinal_features:
    df_clean[f'{feature}_encoded'] = label_encoder.fit_transform(df_clean[feature])
    print(f"  ‚úÖ Label encoded: {feature}")

# 3. Scaling
print("\n3. ‚öñÔ∏è FEATURE SCALING:")
numerical_features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
data_standard = standard_scaler.fit_transform(df_clean[numerical_features])
data_minmax = minmax_scaler.fit_transform(df_clean[numerical_features])
print("  üìä StandardScaler - Mean ~ 0, Std ~ 1")
print(f"  üìä MinMaxScaler - Range: [{data_minmax.min():.2f}, {data_minmax.max():.2f}]")
scaler = StandardScaler()
df_clean[numerical_features] = scaler.fit_transform(df_clean[numerical_features])
print("  ‚úÖ Applied StandardScaler to numerical features")
print("\nüìà Transformed Data Summary:")
display(df_clean[numerical_features].describe())

# =============================================================================
# PHASE 4: DATA REDUCTION
# =============================================================================

print("\n" + "=" * 60)
print("PHASE 4: DATA REDUCTION")
print("=" * 60)

# Prepare features and target
feature_columns = [col for col in df_clean.columns if col not in ['Outcome', 'AgeGroup', 'BMICategory', 'GlucoseCategory'] and not col.endswith('_encoded')]
X = df_clean[feature_columns]
y = df_clean['Outcome']
print(f"üìä Features for reduction: {feature_columns}")

# 1. Feature Selection
print("\n1. üîç FEATURE SELECTION:")
plt.figure(figsize=(12, 8))
correlation_matrix = X.corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', mask=mask)
plt.title('Feature Correlation Matrix (Upper Triangle)')
plt.tight_layout()
plt.show()
selector = SelectKBest(score_func=mutual_info_classif, k='all')
selector.fit(X, y)
feature_scores = pd.DataFrame({
    'Feature': feature_columns,
    'Score': selector.scores_
}).sort_values('Score', ascending=False)
print("üìã Feature Importance Scores (Mutual Information):")
display(feature_scores)
plt.figure(figsize=(10, 6))
sns.barplot(data=feature_scores, x='Score', y='Feature', palette='viridis')
plt.title('Feature Importance Scores (Mutual Information)')
plt.xlabel('Mutual Information Score')
plt.tight_layout()
plt.show()

# 2. Dimensionality Reduction (PCA)
print("\n2. üìâ DIMENSIONALITY REDUCTION (PCA):")
pca = PCA()
X_pca = pca.fit_transform(X)
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.6, color='skyblue', label='Individual')
ax1.step(range(1, len(cumulative_variance) + 1), cumulative_variance, where='mid', label='Cumulative', color='red')
ax1.axhline(y=0.95, color='r', linestyle='--', label='95% Variance')
ax1.set_xlabel('Principal Components')
ax1.set_ylabel('Explained Variance Ratio')
ax1.set_title('PCA - Scree Plot')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax2.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, 'bo-', linewidth=2, markersize=6)
ax2.axhline(y=0.95, color='r', linestyle='--', label='95% Variance')
ax2.set_xlabel('Number of Components')
ax2.set_ylabel('Cumulative Explained Variance')
ax2.set_title('Cumulative Explained Variance')
ax2.grid(True, alpha=0.3)
ax2.legend()
plt.tight_layout()
plt.show()
optimal_components = np.argmax(cumulative_variance >= 0.95) + 1
print(f"üéØ Optimal number of components for 95% variance: {optimal_components}")
print(f"üìä Explained variance with {optimal_components} components: {cumulative_variance[optimal_components-1]:.3f}")
print("\nüìà Explained variance by component:")
for i, (individual, cumulative) in enumerate(zip(explained_variance, cumulative_variance), 1):
    print(f"  PC{i}: Individual={individual:.3f}, Cumulative={cumulative:.3f}")

# =============================================================================
# PHASE 5: DATA IMBALANCE HANDLING (Without imblearn)
# =============================================================================

print("\n" + "=" * 60)
print("PHASE 5: DATA IMBALANCE HANDLING")
print("=" * 60)

# 1. Class Distribution Analysis
target_counts = df_clean['Outcome'].value_counts()
imbalance_ratio = target_counts[0] / target_counts[1]
print(f"üìä Class Distribution: {dict(target_counts)}")
print(f"‚öñÔ∏è Imbalance Ratio: {imbalance_ratio:.2f}:1")
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
bars = plt.bar(target_counts.index, target_counts.values, color=['skyblue', 'lightcoral'])
plt.title('Class Distribution')
plt.xlabel('Outcome (0: No Diabetes, 1: Diabetes)')
plt.ylabel('Count')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height, f'{height}', ha='center', va='bottom')
plt.subplot(1, 2, 2)
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90)
plt.title('Class Distribution (%)')
plt.tight_layout()
plt.show()

# 2. Manual Balancing Techniques (Without imblearn)
print("\n2. ‚öñÔ∏è BALANCING TECHNIQUES (Manual Implementation):")

# Method 1: Random Oversampling of minority class
def manual_random_oversample(X, y):
    from sklearn.utils import resample
    # Combine features and target
    data = pd.concat([X, y], axis=1)
    
    # Separate majority and minority classes
    majority_class = data[data['Outcome'] == 0]
    minority_class = data[data['Outcome'] == 1]
    
    # Oversample minority class
    minority_oversampled = resample(minority_class,
                                  replace=True,     # sample with replacement
                                  n_samples=len(majority_class),  # to match majority class
                                  random_state=42)
    
    # Combine majority class with oversampled minority class
    balanced_data = pd.concat([majority_class, minority_oversampled])
    
    # Shuffle the dataset
    balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)
    
    # Separate features and target
    X_balanced = balanced_data.drop('Outcome', axis=1)
    y_balanced = balanced_data['Outcome']
    
    return X_balanced, y_balanced

# Apply manual oversampling
X_balanced, y_balanced = manual_random_oversample(X, y)
balanced_counts = y_balanced.value_counts()
print(f"üîÑ After Manual Random Oversampling - Class Distribution: {dict(balanced_counts)}")

# Visualize balanced data
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
bars_before = plt.bar(target_counts.index, target_counts.values, color=['skyblue', 'lightcoral'], alpha=0.7)
plt.title('Class Distribution - Before Balancing')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.subplot(1, 2, 2)
bars_after = plt.bar(balanced_counts.index, balanced_counts.values, color=['lightgreen', 'lightgreen'], alpha=0.7)
plt.title('Class Distribution - After Manual Oversampling')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

print("\nüìã Available balancing techniques implemented:")
print("  1. Manual Random Oversampling (Implemented)")
print("  2. Manual Undersampling (Can be implemented similarly)")
print("  3. Custom SMOTE-like approach (Advanced)")

# =============================================================================
# FINAL RESULTS & DATA EXPORT
# =============================================================================

print("\n" + "=" * 60)
print("FINAL RESULTS & DATA EXPORT")
print("=" * 60)

# Save processed data
output_filename = 'diabetes_processed.csv'
df_clean.to_csv(output_filename, index=False)
print(f"üíæ Processed data saved as: {output_filename}")

# Save balanced dataset
balanced_data = pd.concat([X_balanced, y_balanced], axis=1)
balanced_data.to_csv('diabetes_balanced.csv', index=False)
print(f"üíæ Balanced data saved as: diabetes_balanced.csv")

# Create and save data dictionary
data_dict = {
    'Feature': [],
    'Description': [],
    'Type': [],
    'Range/Values': []
}
original_features = {
    'Pregnancies': 'Number of times pregnant',
    'Glucose': 'Plasma glucose concentration (mg/dL)',
    'BloodPressure': 'Diastolic blood pressure (mm Hg)',
    'SkinThickness': 'Triceps skin fold thickness (mm)',
    'Insulin': '2-Hour serum insulin (mu U/ml)',
    'BMI': 'Body mass index (kg/m¬≤)',
    'DiabetesPedigreeFunction': 'Diabetes pedigree function',
    'Age': 'Age in years',
    'Outcome': 'Target variable (0: No diabetes, 1: Diabetes)'
}
engineered_features = {
    'AgeGroup': 'Age categories (Young, Middle-aged, Senior, Elderly)',
    'BMICategory': 'BMI categories (Underweight, Normal, Overweight, Obese)',
    'GlucoseCategory': 'Glucose level categories (Low, Normal, Prediabetes, Diabetes, Severe Diabetes)',
    'AgeGroup_encoded': 'Encoded age groups (0: Young, 1: Middle-aged, 2: Senior, 3: Elderly)',
    'BMICategory_encoded': 'Encoded BMI categories (0: Underweight, 1: Normal, 2: Overweight, 3: Obese)',
    'GlucoseCategory_encoded': 'Encoded glucose categories (0: Low, 1: Normal, 2: Prediabetes, 3: Diabetes, 4: Severe Diabetes)'
}
for feature, description in original_features.items():
    if feature in df_clean.columns:
        data_dict['Feature'].append(feature)
        data_dict['Description'].append(description)
        data_dict['Type'].append('Numerical')
        if feature == 'Outcome':
            data_dict['Range/Values'].append('0, 1')
        else:
            data_dict['Range/Values'].append(f"{df_clean[feature].min():.2f} - {df_clean[feature].max():.2f}")
for feature, description in engineered_features.items():
    if feature in df_clean.columns:
        data_dict['Feature'].append(feature)
        data_dict['Description'].append(description)
        if '_encoded' in feature:
            data_dict['Type'].append('Numerical (Encoded)')
            data_dict['Range/Values'].append(f"{df_clean[feature].min():.0f} - {df_clean[feature].max():.0f}")
        else:
            data_dict['Type'].append('Categorical')
            unique_vals = df_clean[feature].unique()
            data_dict['Range/Values'].append(str(list(unique_vals)))
data_dict_df = pd.DataFrame(data_dict)
data_dict_df.to_csv('data_dictionary.csv', index=False)
print("üìñ Data dictionary saved as: data_dictionary.csv")

# Display final dataset info
print(f"\nüéâ PREPROCESSING PIPELINE COMPLETED SUCCESSFULLY!")
print(f"üìä Final dataset shape: {df_clean.shape}")
print(f"üî¢ Number of features: {len(df_clean.columns)}")
print(f"üë• Number of samples: {len(df_clean)}")
print(f"üéØ Target distribution: {dict(df_clean['Outcome'].value_counts())}")
print(f"üìà Balanced dataset shape: {balanced_data.shape}")
print(f"üéØ Balanced target distribution: {dict(balanced_data['Outcome'].value_counts())}")

print("\nüìã FINAL DATASET SUMMARY:")
print(df_clean.info())

print("\nüëÄ First 3 rows of processed data:")
display(df_clean.head(3))

print("\nüìÅ FILES GENERATED:")
print("  1. diabetes_processed.csv - Cleaned and transformed dataset")
print("  2. diabetes_balanced.csv - Balanced dataset for modeling")
print("  3. data_dictionary.csv - Comprehensive feature documentation")
```
